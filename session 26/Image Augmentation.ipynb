```python
path = "./32a50d167d1440d49e23c648dbd9bcd8.jpg"
```

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing import image
```

```python
img = image.load_img(path, target_size=(50,50))
img = image.img_to_array(img)/255.
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
img = img.reshape(1, 50,50,3)
```

```python
img.shape
```

```python
datagen = image.ImageDataGenerator(rotation_range=20, 
                         width_shift_range=0.2,
                         height_shift_range=0.2,
                         shear_range=0.2,
                         zoom_range=0.3,
                         horizontal_flip=True)
```

```python
arr = datagen.flow(img, batch_size=1)
```

```python
c = 0
for im in arr:
    if c==10:
        break
    print(im.shape)
    plt.imshow(im[0])
    plt.show()
    c+= 1
```
