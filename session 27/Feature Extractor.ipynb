```python
from keras.layers import Input, Conv2D, MaxPool2D, Dense, Flatten
```

```python
inp = Input(shape = (224,224,3))

conv1 = Conv2D(64, (3,3), activation = "relu")(inp)

maxpool1 = MaxPool2D()(conv1)

conv2 = Conv2D(128, (3,3), activation = "relu")(maxpool1)

maxpool2 = MaxPool2D()(conv2)

ft = Flatten()(maxpool2)

fc1 = Dense(64, activation="relu")(ft)
fc2 = Dense(10, activation="softmax")(fc1)


conv_branch2_1 = Conv2D(64, (3,3), activation='relu')(maxpool1)
maxpool_brach2_1 = MaxPool2D()(conv_branch2_1)
ft_branch2 = Flatten()(maxpool_brach2_1)
output2 = Dense(4, activation="linear")(ft_branch2)
```

```python
from keras.models import Model
```

```python
model = Model(inputs = inp, outputs = [fc2, output2])
```
